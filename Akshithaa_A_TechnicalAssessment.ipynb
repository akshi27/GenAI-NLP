```python
import pandas as pd

# Load the dataset
df_rbi = pd.read_csv(
    '/content/HBS Table No. 43 _ Major Monetary Policy Rates and Reserve Requirements - Bank Rate, LAF (Repo, Revers.csv'
)

# Rename columns for consistency
df_rbi.columns = [
    'Effective Date',
    'Bank Rate',
    'Repo Rate',
    'Reverse Repo Rate',
    'SDF Rate',
    'MSF Rate',
    'CRR',
    'SLR'
]

# Convert Effective Date to datetime
df_rbi['Effective Date'] = pd.to_datetime(df_rbi['Effective Date'], format='%d-%m-%Y', errors='coerce')

# Drop rows with invalid dates
initial_len = len(df_rbi)
df_rbi.dropna(subset=['Effective Date'], inplace=True)
dropped_rows = initial_len - len(df_rbi)

# Add Month column (period for grouping)
df_rbi['Month'] = df_rbi['Effective Date'].dt.to_period('M').dt.to_timestamp()

# Function to merge rows by month using median
def merge_monthly(group):
    result = {'Month': group['Month'].iloc[0]}
    for col in group.columns:
        if col in ['Month', 'Effective Date']:
            continue
        numeric_vals = pd.to_numeric(group[col], errors='coerce').dropna()
        if numeric_vals.empty:
            result[col] = None
        elif len(numeric_vals.unique()) == 1:
            result[col] = numeric_vals.iloc[0]
        else:
            result[col] = numeric_vals.median()
    return pd.Series(result)

# Group by Month and apply merge logic
df_rbi = df_rbi.groupby('Month').apply(merge_monthly).reset_index(drop=True)

# Convert Month to MM-YYYY format
df_rbi['Month'] = pd.to_datetime(df_rbi['Month']).dt.strftime('%m-%Y')

print(df_rbi)
```

```python
# Define columns to drop due to high missingness
cols_to_drop = ['SDF Rate', 'MSF Rate', 'SLR']

# Drop the sparse columns
df_rbi = df_rbi.drop(columns=cols_to_drop)

# Forward-fill remaining numeric columns
df_rbi.fillna(method='ffill', inplace=True)

# Final NaN diagnostics
nan_ratios = df_rbi.isna().mean().round(3)
print("NaN Ratio per Column (After Cleaning):")
print(nan_ratios)

# Preview cleaned DataFrame
print("\nCleaned RBI Monthly Data:")
print(df_rbi)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import grangercausalitytests

variables = [
    'CPI', 'PPI', 'Bank Rate', 'Repo Rate', 'Reverse Repo Rate', 'CRR',
    'Exchange Rate', 'Rural Index',
    'Rural Inflation (%)', 'Urban Index', 'Combined Index', 'Combined Inflation (%)',
    'GDP growth rate', 'Consumer Confidence Index', 'Gasoline Price (USD/Litre)'
]

max_lag = 3  # test up to lag 3 months
gc_results = {}

for var in variables:
    print(f"\n⏳ Testing {var} → Closing Price")
    try:
        test_result = grangercausalitytests(
            model_df[[var, 'Closing Price']], maxlag=max_lag, verbose=False
        )
        p_values = [round(test_result[i+1][0]['ssr_ftest'][1], 4) for i in range(max_lag)]
        gc_results[var] = p_values
        print(f"Lag-wise p-values: {p_values}")
    except Exception as e:
        print(f"⚠️ Could not test {var}: {e}")
```

##Granger Causality tests

```python
# Show min p-value for each variable

granger_ranked = {k: min(v) for k, v in gc_results.items()}
granger_ranked = dict(sorted(granger_ranked.items(), key=lambda item: item[1]))

print("\nPredictive Variables by Granger p-value (lower is better):")
for var, pval in granger_ranked.items():
    print(f"{var}: min p = {pval}")
```

## XG Boost + L1, L2 Regularization + Time features Stock Predictor (Best)


```python
model_df_c = model_df.copy()

# Convert 'Month' to datetime
model_df_c['Month_dt'] = pd.to_datetime(model_df_c['Month'], format="%m-%Y")

# Add Month Number (1 to 12)
model_df_c['Month_Num'] = model_df_c['Month_dt'].dt.month

# Add Year
model_df_c['Year'] = model_df_c['Month_dt'].dt.year

# Add Quarter (optional)
model_df_c['Quarter'] = model_df_c['Month_dt'].dt.quarter

# Add continuous TimeIndex (0 for earliest month, 1, 2, ...)
model_df_c = model_df_c.sort_values('Month_dt').reset_index(drop=True)
model_df_c['TimeIndex'] = range(len(model_df_c))

# Drop helper column
model_df_c.drop(columns=['Month_dt'], inplace=True)
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Step 2: Split features and target
X = model_df_c.drop(columns=['Month', 'Closing Price'])
y = model_df_c['Closing Price']

# Step 3: Train/Val/Test split: 80/10/10
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)
# 0.1111 ≈ 10 / 90 to get 10% of original as validation

# Step 4: Prepare DMatrix for XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)
dtest = xgb.DMatrix(X_test, label=y_test)

# Step 5: Define model parameters
params = {
    'objective': 'reg:squarederror',
    'learning_rate': 0.05,
    'max_depth': 2,
    'eval_metric': 'rmse',
    'lambda': 1.0,   # L2 regularization term
    'alpha': 0.5,    # L1 regularization term
    'seed': 42
}

# Step 6: Train with early stopping
model = xgb.train(
    params,
    dtrain,
    num_boost_round=100,
    evals=[(dtrain, 'train'), (dval, 'val')],
    early_stopping_rounds=6,
    verbose_eval=False
)

# Step 7: Evaluation function
def evaluate(model, dmatrix, y_true, label):
    preds = model.predict(dmatrix)
    r2 = r2_score(y_true, preds)
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))

    print(f"\n{label} Set Performance:")
    print(f"R²: {r2:.4f} | MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# Step 8: Evaluate on train, val, test
evaluate(model, dtrain, y_train, "Train")
evaluate(model, dval, y_val, "Validation")
evaluate(model, dtest, y_test, "Test")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Step 1: Generate predictions for the test set
y_pred = model.predict(dtest)
y_true = y_test.reset_index(drop=True)
y_pred = pd.Series(y_pred)

# Step 2: Plot actual vs predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_true, y=y_pred, color='dodgerblue', s=60, alpha=0.7)

# Ideal diagonal line
min_val = min(y_true.min(), y_pred.min())
max_val = max(y_true.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], '--', color='red')

plt.xlabel("Actual Scaled Closing Price")
plt.ylabel("Predicted Scaled Closing Price")
plt.title("Predicted vs Actual Stock Price (XGBoost)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##XG Boost + Time features + Bagging

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Step 1: Prepare features and target
X = model_df_c.drop(columns=['Month', 'Closing Price'])
y = model_df_c['Closing Price']

# Step 2: Train/Val/Test split (80/10/10)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)

# Step 3: Define XGBoost base model
xgb_base = XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.05,
    max_depth=2,
    reg_lambda=1.0,  # L2
    reg_alpha=0.5,   # L1
    n_estimators=100,
    random_state=42
)

# Step 4: Wrap with BaggingRegressor
bagged_model = BaggingRegressor(
    estimator=xgb_base,
    n_estimators=10,         # Number of bagging rounds
    max_samples=0.8,         # Bootstrap sample size (80%)
    bootstrap=True,
    n_jobs=-1,
    random_state=42
)

# Step 5: Fit the model
bagged_model.fit(X_train, y_train)

# Step 6: Evaluation function
def evaluate(model, X, y, label):
    preds = model.predict(X)
    r2 = r2_score(y, preds)
    mae = mean_absolute_error(y, preds)
    rmse = np.sqrt(mean_squared_error(y, preds))
    print(f"\n{label} Set Performance:")
    print(f"R²: {r2:.4f} | MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# Step 7: Evaluate
evaluate(bagged_model, X_train, y_train, "Train")
evaluate(bagged_model, X_val, y_val, "Validation")
evaluate(bagged_model, X_test, y_test, "Test")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Step 1: Split features and target
X = model_df_c.drop(columns=['Month', 'Closing Price'])
y = model_df_c['Closing Price']

# Step 2: Train/Val/Test split: 80/10/10
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1111, random_state=42)
# (0.1111 ≈ 10% of 90% to get 10% validation)

# Step 3: Prepare Dataset for LightGBM
dtrain = lgb.Dataset(X_train, label=y_train)
dval = lgb.Dataset(X_val, label=y_val, reference=dtrain)

# Step 4: Define model parameters
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'learning_rate': 0.05,
    'max_depth': 2,
    'lambda_l2': 1.0,
    'lambda_l1': 0.5,
    'seed': 42,
    'verbosity': -1
}

# Step 5: Train model with early stopping and silent logging
model = lgb.train(
    params,
    dtrain,
    num_boost_round=100,
    valid_sets=[dtrain, dval],
    valid_names=['train', 'val'],
    callbacks=[
        lgb.early_stopping(6),
        lgb.log_evaluation(0)  # set to >0 like 10 for periodic logs
    ]
)

# Step 6: Define evaluation function
def evaluate(model, X, y_true, label):
    preds = model.predict(X, num_iteration=model.best_iteration)
    r2 = r2_score(y_true, preds)
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))

    print(f"\n📊 {label} Set Performance:")
    print(f"R²: {r2:.4f} | MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# Step 7: Evaluate
evaluate(model, X_train, y_train, "Train")
evaluate(model, X_val, y_val, "Validation")
evaluate(model, X_test, y_test, "Test")
```

###Signal 1: AVMD (Adjusted Volume-Momentum Divergence)

```python
df_signal['AVMD'] = (
    df_signal['Return'].rolling(3).std() /
    df_signal['Volume'].ewm(span=3, adjust=False).mean()
)
df_signal['AVMD'] = (df_signal['AVMD'] - df_signal['AVMD'].mean()) / df_signal['AVMD'].std()
```

###Signal 2: EWIPI (Exchange-Weighted Input Pressure Index)

```python
rolling_fx = df_signal['Exchange Rate'].rolling(6).mean()
ewipi_raw = np.log1p(df_signal['PPI'] * df_signal['Gasoline Price (USD/Litre)']) * (
    df_signal['Exchange Rate'] / rolling_fx
)
df_signal['EWIPI'] = -((ewipi_raw - ewipi_raw.mean()) / ewipi_raw.std())
```

###Signal 3: DSES (Demand-Sentiment Elasticity Score)

```python
dses_raw = (
    df_signal['Consumer Confidence Index'] * (df_signal['GDP growth rate'] ** 1.5)
) / (
    (df_signal['Bank Rate'] ** 1.2) * (df_signal['Urban Inflation (%)'] ** 0.8)
)
df_signal['DSES'] = (dses_raw - dses_raw.mean()) / dses_raw.std()
```

```python
# Final Composite Signal
df_signal['CompositeSignal'] = df_signal[['AVMD', 'EWIPI', 'DSES']].mean(axis=1)
```

```python
# Defining threshold
threshold = 0.5

# Step 2: Generate trading positions
df_signal['Position'] = 0  # neutral default
df_signal.loc[df_signal['CompositeSignal'] > threshold, 'Position'] = 1   # long/buy
df_signal.loc[df_signal['CompositeSignal'] < -threshold, 'Position'] = -1  # short/sell
```

##Model 1 (Long-term trading)

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error

# ---- Step 1: Prepare Data ---- #
features = ['AVMD', 'EWIPI', 'DSES']
target = 'Return'

# Prepare DataFrame
df_ml = df_signal.dropna(subset=features + [target]).copy()
df_ml['Return'] = df_ml['Return'] / 100.0

# Target = next month's return
df_ml['TargetReturn'] = df_ml['Return'].shift(-1)
df_ml.dropna(subset=['TargetReturn'], inplace=True)

# ---- Step 2: Train Random Forest ---- #
X = df_ml[features]
y = df_ml['TargetReturn']
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

rf = RandomForestRegressor(n_estimators=100, max_depth=4, random_state=42)
rf.fit(X_train, y_train)

# Predict and smooth
df_ml['PredictedReturn'] = rf.predict(X) / 100.0  # decimal

# Smoothing
df_ml['PredictedReturnSmoothed'] = df_ml['PredictedReturn'].rolling(window=3).mean()

# Volatility filter
df_ml['Volatility'] = df_ml['Return'].rolling(window=3).std()
vol_threshold = df_ml['Volatility'].quantile(0.7)
low_vol_mask = df_ml['Volatility'] < vol_threshold

# Scaled positions (cap extreme predictions)
threshold = 0.0028
scaled_pred = df_ml['PredictedReturnSmoothed'] / threshold
scaled_pred = scaled_pred.clip(-1, 1)

# Only trade during low volatility periods
df_ml['ScaledPosition'] = 0
df_ml.loc[low_vol_mask, 'ScaledPosition'] = scaled_pred[low_vol_mask]

# Strategy returns
df_ml['StrategyReturn'] = df_ml['ScaledPosition'].shift(1) * df_ml['Return']
df_ml['StrategyReturn'] = df_ml['StrategyReturn'].clip(lower=-0.10)  # Stop-loss simulation

# Filter out extreme strategy returns
df_ml = df_ml[df_ml['StrategyReturn'].abs() < 1]

# Cumulative returns
df_ml['Cumulative Strategy Return'] = (1 + df_ml['StrategyReturn']).cumprod()
df_ml['Cumulative Buy & Hold'] = (1 + df_ml['Return']).cumprod()

# Sharpe Ratio
sharpe = (df_ml['StrategyReturn'].mean() / df_ml['StrategyReturn'].std()) * np.sqrt(12)

# Max Drawdown
rolling_max = df_ml['Cumulative Strategy Return'].cummax()
drawdown = df_ml['Cumulative Strategy Return'] / rolling_max - 1
max_drawdown = drawdown.min()

# Turnover
df_ml['Turnover'] = df_ml['ScaledPosition'].diff().abs()
monthly_turnover = df_ml['Turnover'].mean()
estimated_daily_turnover = monthly_turnover / 22

# ---- Annualized Return ---- #
n_months = len(df_ml)
final_cum_return = df_ml['Cumulative Strategy Return'].iloc[-1]
annualized_return = final_cum_return**(12 / n_months) - 1

# ---- Final Results ---- #
print("\nML Strategy Backtest:")
print(f"Sharpe Ratio: {sharpe:.2f}")
print(f"Max Drawdown: {max_drawdown:.2%}")
print(f"Annualized Return: {annualized_return:.2%}")
print(f"Monthly Turnover: {monthly_turnover:.2%}")
print(f"Estimated Daily Turnover: {estimated_daily_turnover:.2%}")
```

```python
import numpy as np

best_sharpe = -np.inf
best_threshold = None
best_stats = {}

threshold_range = np.linspace(0.001, 0.05, 50)  # try thresholds from 0.1% to 5%

for threshold in threshold_range:
    df_temp = df_ml.copy()

    # Apply thresholding logic
    df_temp['Position'] = 0
    df_temp.loc[df_temp['PredictedReturn'] > threshold, 'Position'] = 1
    df_temp.loc[df_temp['PredictedReturn'] < -threshold, 'Position'] = -1

    # Calculate shifted strategy return
    df_temp['StrategyReturn'] = df_temp['Position'].shift(1) * df_temp['Return']
    df_temp['Turnover'] = df_temp['Position'].diff().abs()

    # Cumulative returns for drawdown and annualized return
    df_temp['CumulativeReturn'] = (1 + df_temp['StrategyReturn']).cumprod()
    rolling_max = df_temp['CumulativeReturn'].cummax()
    drawdown = df_temp['CumulativeReturn'] / rolling_max - 1
    max_drawdown = drawdown.min()

    # Risk metrics
    mean_ret = df_temp['StrategyReturn'].mean()
    std_ret = df_temp['StrategyReturn'].std()
    sharpe = mean_ret / std_ret * np.sqrt(12) if std_ret > 0 else 0
    turnover = df_temp['Turnover'].mean()
    estimated_daily_turnover = turnover / 21

    total_periods = df_temp.shape[0]
    annualized_return = (df_temp['CumulativeReturn'].iloc[-1]) ** (12 / total_periods) - 1 if total_periods > 0 else 0

    # Save if turnover is valid and Sharpe is better
    if turnover < 0.40 and sharpe > best_sharpe:
        best_sharpe = sharpe
        best_threshold = threshold
        best_stats = {
            'Sharpe': sharpe,
            'Turnover': turnover,
            'Threshold': threshold,
            'MaxDrawdown': max_drawdown,
            'AnnualizedReturn': annualized_return,
            'EstimatedDailyTurnover': estimated_daily_turnover
        }

# Print the best result
print("Optimal Threshold Found:")
print(f"Threshold: {best_threshold:.4f}")
print(f"Sharpe Ratio: {best_stats['Sharpe']:.2f}")
print(f"Avg Monthly Turnover: {best_stats['Turnover']:.2%}")
print(f"Estimated Daily Turnover: {best_stats['EstimatedDailyTurnover']:.2%}")
print(f"Max Drawdown: {best_stats['MaxDrawdown']:.2%}")
print(f"Annualized Return: {best_stats['AnnualizedReturn']:.2%}")
```
